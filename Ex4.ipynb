```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Position_Salaries.csv")
data
```

```python
x=data.iloc[:,1:2].values
x
```

```python
y=data.iloc[:,2:3].values
y
```

```python
from sklearn.linear_model import LinearRegression as LR
lin_reg=LR()
lin_reg.fit(x,y)
```

```python
from  sklearn.preprocessing import PolynomialFeatures as PF
poly_reg2=PF(degree=2)
x_poly=poly_reg2.fit_transform(x)
#Linear Regression
lin_reg2=LR()
lin_reg2.fit(x_poly,y)
```

```python
poly_reg3=PF(degree=3)
x_poly3=poly_reg3.fit_transform(x)
lin_reg3=LR()
lin_reg3.fit(x_poly3,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
plt.title('Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg2.predict(poly_reg2.fit_transform(x)),color='blue')
plt.plot(x,lin_reg3.predict(poly_reg3.fit_transform(x)),color='green')
plt.title('Polynomial Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg2.predict(x_poly),color='blue')
plt.plot(x,lin_reg3.predict(x_poly3),color='green')
plt.title('Polynomial Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
